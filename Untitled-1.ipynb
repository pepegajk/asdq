Шаг 1. Импортируем pandas для работы с датасетом и matplotlib для работы с графиками

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
dota2 = pd.read_csv('Dota2_Heroes_by_MMR_Bracket.csv')
dota2
```

Шаг 2. Сделаем 3 штуки линейного графика

```python
plt.figure(figsize=(15, 5))
plt. plot(dota2['Heroes'].head(10), dota2['(<2K MMR) PickRate(%)'].head(10), label='пикрейт')
plt.title('пикрейт2к')
plt.xlabel('Имя персонажа')
plt.ylabel('Рост')
plt.legend()
plt.show()
```

Был сделан линейный график на котором отображен пикрейт первых десяти персонажей

```python
plt.figure(figsize=(15, 5))
plt. plot(dota2['Heroes'].head(10), dota2['(<2K MMR) WinRate(%)'].head(10), label='винрейт', marker = 'o', markersize = 7)
plt.title('винтрейт2к')
plt.xlabel('Имя персонажа')
plt.ylabel('винрейт')
plt.legend()
plt.show()
```

Был сделан линейный график на котором отображен пикрейт персонажей на 2к

```python
plt.figure(figsize=(15, 5))
plt. plot(dota2['Heroes'].head(10), dota2['(>2K MMR) KDA Ratio'].head(10), label='кд', marker = 'o', markersize = 7, color = 'pink')
plt.title('кд персонажей 2к')
plt.xlabel('Имя персонажа')
plt.ylabel('кд')
plt.legend()
plt.grid()
plt.show()
```
